```python
import numpy as np
import os,glob,sys
```

```python
old_manual_positions = [[-5107.8,-1999.3],
                        [444.55,1020.75],]

new_manual_positions = [[-5103.5, -2000.3],
                        [446, 1019.0], ]

shifts = np.array(new_manual_positions) - np.array(old_manual_positions)
print(shifts)
# calculate mean
manual_shift = np.round(np.mean(shifts, axis=0), 2)
print(manual_shift)
```

```python
data_folder = r'D:\Pu\20220209-P_brain_CTP11-1000_CTP12_from_0204'
old_positions = np.loadtxt(os.path.join(data_folder, 'adjusted_translated_positions_all.txt'), delimiter=',')
```

```python
print(old_positions)
```

```python
adjusted_positions = old_positions + manual_shift
```

```python
adj_save_filename = os.path.join(data_folder, 'double_adjusted_translated_positions_all.txt')
print(adj_save_filename)
np.savetxt(adj_save_filename, adjusted_positions, fmt='%.2f', delimiter=',')
```
